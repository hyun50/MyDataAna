```python
#1차 함수 그리기
x = range(0,30)
y = [v*v for v in  x]

plt.plot(x, y, 'ro')
plt.show()
```

```python
#2차 함수 그리기, 파란색으로 점찍기
x = range(0,30)
y = [v*v for v in  x]

plt.plot(x, y, 'bo')
plt.show()
```

```python
#2차 함수 기울기 크게 그리기, 파란색으로 점찍기
x = range(0,30)
y = [v*v*2 for v in  x]

plt.plot(x, y, 'bo')
plt.show()
```

```python
#그래프 각각 그리기
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

x1 = np.array([0,1,2,3])
y1 = np.array([13,18,11,10])

x2 = np.array([0,1,2,3])
y2 = np.array([32, 81, 11, 11])

ax1.plot(x1, y1)
ax2.plot(x2, y2)
plt.show()
```

```python
#그래프 각각 그리기 2  ; 여기는 이해가 잘 가지 않음.
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) #행열, 2행 2열중 1번째를 의미함
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

x1 = np.array([0,1,2,3])
y1 = np.array([13,18,11,10])

x2 = np.array([0,1,2,3])
y2 = np.array([32, 81, 11, 11])

x3 = np.array([0,1,2,3])
y3 = np.array([320, 810, 110, 110])

ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.plot(x3, y3)
plt.show()
```
