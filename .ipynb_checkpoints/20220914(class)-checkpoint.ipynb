```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
#2-4 실습
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([100, 200, 300, 400])

# 1행 2열, 두번째
plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("INCOME2")

#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([130, 80, 10, 100])

# 1행 2열, 두번째
plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("SALES2")

plt.suptitle("MY SHOP")
plt.show()
```

```python
##plt.plot() : 산점도, 선(기본그래프, 선, 산점도)
##plt.scatter() : 산점도
##plt.bar() : 막대그래프
##plt.barh() : 막대그래프를 수평으로
##plt.hist() : 히스토드램
##plt.pie() : 원그래프
```

```python
plt.figure(figsize=(10,10))

y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
plt.pie(y, labels=addlbl)
```

```python
y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
myexplode = [0.2, 0, 0, 0, 0.3]

plt.pie(y, labels = addlbl, explode = myexplode)
```

```python
y = np.array([30,20,35,10, 5])
addlbl = ['사과', '바나나', '딸기', '수박', '오렌지']
myexplode = [0.2, 0, 0, 0, 0.3]

plt.pie(y, labels = addlbl, explode = myexplode)
plt.legend()
```
