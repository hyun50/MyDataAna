```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# 1행 2열, 첫번째
plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# 1행 2열, 두번째
plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set_style("whitegrid")

tips = sns.load_dataset("tips")  # 인터넷이 켜져 있어야 함.
tips
```

```python
##시본 48개 그래프
##막대그래프(countplot, barplot)
##히스토그램(histplot)
##산점도
##heatmap
##회기직선
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips) #x 범주, Y 수치형
plt.show()
```

```python
#실습, 요일별 인원수(size)
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="size", data=tips) #x 범주, Y 수치형
plt.show()
```

```python
#실습, 요일별 인원수
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="size", data=tips) #x 범주, Y 수치형
plt.title("day and size")
plt.show()
```

```python
### sns.boxplot()
###sns implot()
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
plt.show()
```

```python
fg = sns.load_dataset("flights")
fg.head(5)
fg
```

```python
#피벗테이블을 사용해서 보여줌
fgp = fg.pivot("month", "year", "passengers")
fgp
```

```python
##sns.boxplot()
##sns.impolt()
##sns.neatmap
plt.figure(figsize=(10,8))
sns.heatmap(fgp)
plt.show()
```

```python
##sns.boxplot()
##sns.impolt()
##sns.neatmap()
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt='d')
plt.show()
```
